```python
# column summary
def column_summary(df):
    summary = []
    for col in df.columns:
        data_type = df[col].dtype
        missing = df[col].isnull().mean() * 100
        unique = df[col].nunique()
        sample_values = df[col].dropna().unique()[:10]
        summary.append({
            'Column Name': col,
            'Data Type': data_type,
            'Missing Values (%)': round(missing, 2),
            'Unique Values': unique,
            'Sample Values': sample_values,
        })
    return pd.DataFrame(summary)

# Generate and display the summary
summary_df = column_summary(df)
print(summary_df)
```

```python
def clean_text(text):
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Remove special characters
    text = re.sub(r'\s+', ' ', text)  # Remove whitespace
    return text.strip()

df['CUSTOMER_VERBATIM'] = df['CUSTOMER_VERBATIM'].apply(clean_text)
df['CORRECTION_VERBATIM'] = df['CORRECTION_VERBATIM'].apply(clean_text)

df.head(2)
```

```python
# topic sepeartion using wordnet library
from nltk.corpus import wordnet as wn

# Your list of words
words = ['12624', 'advise', 'apart', 'approved', 'assembly', 'auth', 'authorization',
         'avant', 'bezel', 'bubbling', 'button', 'changer', 'check', 'clicking', 'coming',
         'control', 'cover', 'cruise', 'cust', 'customer', 'diagnosis', 'does', 'doesnt',
         'driver', 'driving', 'faire', 'faut', 'finish', 'gets', 'hand', 'heat', 'heated',
         'history', 'horn', 'hot', 'immediately', 'ing', 'inop', 'inspect', 'install',
         'installed', 'leather', 'left', 'light', 'loose', 'making', 'mike', 'moulding',
         'new', 'noise', 'operation', 'order', 'ordered', 'parts', 'peeling', 'piece',
         'pra', 'pre', 'preauthorisation', 'preauthorization', 'previous', 'rear',
         'refer', 'reference', 'repair', 'replace', 'replacement', 'report', 'request',
         'retrofit', 'right', 'seated', 'section', 'securepra', 'separatinginstall',
         'service', 'smoking', 'sop', 'special', 'stance', 'state', 'states', 'steer',
         'steering', 'stitching', 'super', 'tech', 'times', 'trim', 'turn', 'turns',
         'undone', 'une', 'vehicle', 'wheel', 'wont', 'work', 'worked', 'working', 'wrap']

# Function to get the hypernym/topic
def get_word_topic(word):
    synsets = wn.synsets(word)
    if synsets:
        hypernyms = synsets[0].hypernyms()
        if hypernyms:
            return hypernyms[0].lemma_names()[0]  # return first hypernym name
        else:
            return 'generic'
    else:
        return 'unknown'

# Categorize words
word_topics = {word: get_word_topic(word) for word in words}

# Optional: print grouped by topics
from collections import defaultdict

grouped = defaultdict(list)
for word, topic in word_topics.items():
    grouped[topic].append(word)

# Print the grouped topics
for topic, word_list in grouped.items():
    print(f"\n Topic: {topic} ({len(word_list)} words)")
    print(", ".join(word_list))
```
